```python
import pandas as pd
import json
```

```python
# Load CSV file into a DataFrame
file_path = "../data/train.csv"
df = pd.read_csv(file_path)

# Ensure all columns are displayed
pd.set_option("display.max_columns", None)

# Display first few rows
df.head()
```

```python
# Group the dataset by Neighborhood and calculate the average SalePrice
neighborhood_prices = df.groupby("Neighborhood")["SalePrice"].mean()

# Normalize the average SalePrice values between 0 and 100 using min-max scaling
neighborhood_scaled = (neighborhood_prices - neighborhood_prices.min()) / (neighborhood_prices.max() - neighborhood_prices.min()) * 100

# Print the normalized neighborhood price values
print(neighborhood_scaled)
```

```python
# Group the dataset by MiscFeature and calculate the average SalePrice for each feature
misc_feature_prices = df.groupby("MiscFeature")["SalePrice"].mean()

# Normalize the average SalePrice values between 0 and 100 using min-max scaling
misc_feature_scaled = (misc_feature_prices - misc_feature_prices.min()) / (misc_feature_prices.max() - misc_feature_prices.min()) * 100

# Print the normalized MiscFeature price values
print(misc_feature_scaled)
```

```python
# Group the dataset by BldgType and calculate the average SalePrice for each building type
bldg_type_prices = df.groupby("BldgType")["SalePrice"].mean()

# Normalize the average SalePrice values between 0 and 100 using min-max scaling
bldg_type_scaled = (bldg_type_prices - bldg_type_prices.min()) / (bldg_type_prices.max() - bldg_type_prices.min()) * 100

# Print the normalized BldgType price values
print(bldg_type_scaled)
```
