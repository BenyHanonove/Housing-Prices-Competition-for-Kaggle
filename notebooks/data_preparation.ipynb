```python
import pandas as pd
import json
```

```python
# Load CSV file into a DataFrame
file_path = "../data/train.csv"
df = pd.read_csv(file_path)

# Ensure all columns are displayed
pd.set_option("display.max_columns", None)

# Display first few rows
df.head()
```

```python
df.info()
```

```python
# Define the JSON file path
json_path = "../jsons/feature_mappings.json"

# Load the JSON mapping
with open(json_path, "r") as f:
    feature_mappings = json.load(f)

# Define the columns to convert using the JSON file
keys = feature_mappings.keys()

# Create a copy of the original DataFrame to store the normalized values
normalized_df = df.copy()

# Convert columns in DataFrame using the feature JSON mapping
for key in keys:
    # Fill NaN values explicitly before mapping
    normalized_df[key] = normalized_df[key].fillna("NA")  # Ensure NaN is replaced with "NA"
    normalized_df[key] = normalized_df[key].map(feature_mappings[key]) 
```

```python
# Convert CentralAir to binary values
normalized_df["CentralAir"] = normalized_df["CentralAir"].map({"Y": 1, "N": 0})
```

```python
normalized_df.head()
```

```python
# Find if any row has nan as value 
print(normalized_df.isna().sum()[normalized_df.isna().sum() > 0])
```

Handling Missing Values in LotFrontage

The LotFrontage column represents the linear feet of street connected to a property. 
<br/>
Some properties have missing values (NaN), which need to be filled appropriately to maintain data integrity. 
<br/>
There are two common approaches to handling these missing values:
<br/>

Approach 1: Filling NaN with the Overall Median

This method replaces all missing values in LotFrontage with a single median value calculated from the entire dataset. 
<br/>
The median is used instead of the mean because it is less sensitive to extreme values (outliers) and provides a stable estimate.
<br/>
 This is a simple and effective solution but does not account for variations between neighborhoods.
<br/>

Approach 2: Filling NaN with the Median Based on Neighborhood

Since lot sizes often vary depending on the neighborhood, this approach replaces missing values with the median LotFrontage of each respective neighborhood.
<br/>
By grouping the data by Neighborhood, we ensure that missing values are filled using values that are more representative of the area. 
<br/>

This method provides a more localized and precise imputation.

```python
#  Option 1: Fill NaN with the Median (Recommended)
# Since lot sizes vary by neighborhood, using the median value is a robust way to impute missing values:
# normalized_df["LotFrontage"] = normalized_df["LotFrontage"].fillna(normalized_df["LotFrontage"].median())

# Option 2: Fill NaN with the Median Based on Neighborhood
# Since lot sizes depend on the neighborhood, it's better to fill missing values based on the median LotFrontage per neighborhood:
normalized_df["LotFrontage"] = normalized_df.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

Why I Chose Approach 2

I chose to fill NaN values based on the median LotFrontage per neighborhood because lot sizes are not uniform across different areas.
<br/>
Using a single median for the entire dataset may not accurately reflect the typical lot size in each neighborhood.
<br/>
By using the neighborhood-specific median, we ensure that missing values are replaced with values that are more relevant and contextually appropriate,
<br/>
leading to a more accurate dataset.

```python
# Display the 8 rows where MasVnrArea has NaN values
missing_masvnrarea_rows = normalized_df[normalized_df["MasVnrArea"].isna()]

missing_masvnrarea_rows.head(8)
```

Each of the 8 rows where MasVnrArea is NaN does not have a corresponding MasVnrType. 
<br/> 
This means that these houses do not have any masonry veneer applied to their exterior.
<br/> 
Since MasVnrType is missing (which implies no veneer), it is logical to assume that MasVnrArea should also be 0 instead of NaN. 
<br/> 
Therefore, to correct the dataset, I will fill these NaN values in MasVnrArea with 0 to properly reflect that these houses have no masonry veneer.


```python
# Fill NaN values in the MasVnrArea column with 0
normalized_df["MasVnrArea"] = normalized_df["MasVnrArea"].fillna(0)
```

```python
# Filter rows where GarageYrBlt is NaN and GarageType is 0
garage_yr_blt_rows = normalized_df[(normalized_df["GarageYrBlt"].isna()) & (normalized_df["GarageType"] == 0)]

# Print the number of rows that match the condition
print(f"Number of rows where GarageYrBlt is NaN and GarageType is 0: {garage_yr_blt_rows.shape[0]}")
```

The GarageYrBlt column contains null values because these houses do not have a garage (GarageType = 0).
<br/> 
Since a garage was never built for these properties, the GarageYrBlt field remains empty, 
<br/> 
indicating that no construction year exists for a non-existent garage.

```python
# Fill NaN values in the GarageYrBlt column with 0
normalized_df["GarageYrBlt"] = normalized_df["GarageYrBlt"].fillna(0)

# Drop rows where SalePrice is above 500,000
normalized_df = normalized_df[normalized_df['SalePrice'] <= 500000].copy()
```

```python
# Save the normalized DataFrame as a CSV file
normalized_df.to_csv("../data/train_normalize.csv", index=False)

print("CSV file saved successfully at: ../data/train_normalize.csv")
```
